# Machine Learning Classification with Logistic Regression

```python
# Importing modules
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# Data Loading
df = pd.DataFrame(load_breast_cancer().data, columns=load_breast_cancer().feature_names)
df["class"] = load_breast_cancer().target
df
```

```python
# Feature matrix
X = df.drop(["class"], axis=1)

# Target labels
y = df["class"]
```

```python
# Data summary | feature matrix
hist = X.hist(figsize=(15, 15))
```

```python
# Data summary | target labels
y.value_counts().plot.bar()
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Preprocessing
scaler = StandardScaler() # scaler

X_train = scaler.fit_transform(X_train) # fitting and the scaler and scaling the training set

X_test = scaler.transform(X_test) # scaling the test set
```

```python
# Model
model = LogisticRegression()
```

```python
# Training
model.fit(X_train, y_train)
```

```python
# Testing
y_pred = model.predict(X_test)
```

```python
# Performance evaluation
print(classification_report(y_test, y_pred))
```
